Data Click Buy

```python
#Tạo dataframe bằng cách đọc file data click buy
data_CB = pd.read_csv('data_click_buy.csv', index_col=False)
data_CB.isna().sum()
```

```python
data_CB.head()
```

```python
#Xóa đi các cột không cần thiết do không gán index = False khi tạo file csv từ các file trước đó
data_CB = data_CB.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
#Đổi tên cột
data_CB.columns = ['Kích thước màn hình', 'Chipset(CPU)', 'Bộ nhớ trong', 'Ram', 'Độ phân giải Camera', 'Dung lượng pin', 'Tên sản phẩm', 'Giá', 'Hãng sản xuất', 'Ram_', 'Rom_']
```

```python
#Chọn ra các cột và sắp xếp lại thứ tự các cột
selected_column = ['Tên sản phẩm', 'Hãng sản xuất', 'Bộ nhớ trong', 'Ram', 'Chipset(CPU)', 'Kích thước màn hình','Độ phân giải Camera', 'Dung lượng pin', 'Giá']
data_CB = data_CB[selected_column]
```

```python
data_CB.head()
```

Data Điện THOẠI GIÁ KHO(DTGK)

```python
#Tạo dataframe bằng cách đọc file data DTGK
df_DTGK = pd.read_csv('data_DTGK.csv', index_col = False)
df_DTGK.isna().sum()
```

```python
df_DTGK.head()
```

```python
#Xóa đi cột index do khi tạo file data DTGK không có tham số index = False
df_DTGK = df_DTGK.drop(columns=['Unnamed: 0'])
#Đặt lại tên các cột
df_DTGK.columns = ['Kích thước màn hình', 'Chipset(CPU)', 'Bộ nhớ trong', 'Ram', 'Độ phân giải Camera', 'Dung lượng pin', 'Tên sản phẩm', 'Giá', 'Hãng sản xuất']
#Chọn và xếp lại thứ tự cột
selected_column = ['Tên sản phẩm', 'Hãng sản xuất', 'Bộ nhớ trong', 'Ram', 'Chipset(CPU)', 'Kích thước màn hình','Độ phân giải Camera', 'Dung lượng pin', 'Giá']
df_DTGK = df_DTGK[selected_column]
```

```python
df_DTGK.head()
```

Merge Data và Tạo file csv

```python
final_df = pd.concat([data_CB, df_DTGK], ignore_index=True)
```
