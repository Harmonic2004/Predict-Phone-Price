```python
#Lấy ra tất cả các link của sản phẩm
def get_links_from_pages(start_page, end_page):
    driver = webdriver.Chrome()
    all_links = []

    try:
        for page_no in range(start_page, end_page + 1):
            driver.get(f'https://dienthoaigiakho.vn/dien-thoai?page={page_no}')
            sleep(2)  # Đợi trang tải xong
            
            page_source = BeautifulSoup(driver.page_source, 'html.parser')
            samples = page_source.find_all('div', class_='sc-e89d1c58-1 kWDRKa')
            
            for sample in samples:
                product = sample.find('a')
                link = product.get('href')
                if not link.startswith("https://dienthoaigiakho.vn/"):
                    link = "https://dienthoaigiakho.vn/" + link
                all_links.append(link)
                
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")
    
    finally:
        driver.quit()

    return list(set(all_links))


# Sử dụng hàm để lấy liên kết từ trang 1 đến trang 19
links = get_links_from_pages(1, 10)
print(links)
```

```python
print(len(links))
```

```python
#Lấy title
product = []

for link in links:
    driver = webdriver.Chrome()
    driver.get(f'{link}')
    sleep(2)  # Đợi trang tải xong
    
    page_source = BeautifulSoup(driver.page_source, 'html.parser')
    samples = page_source.find('h1', class_='sc-39689109-1 fEvwZo')
    title = samples.text
    product.append((title, link))
```

```python
#Lưu link và title lấy được vào file
def write_links_to_file(links, filename='DTGK_links.txt'):
    with open(filename, 'w', encoding='utf-8') as file:
        for title, link in links:
            title = title.strip()
            link = link.strip()
            file.write(f'{title} : {link}\n')
            
# Ghi các liên kết vào file
write_links_to_file(product)
print("Hoàn thành việc ghi các liên kết vào file.")
```

```python
#Hàm lấy ra chỉ số sản phẩm từ việc test
def get_phone_details(url):
    # Khởi tạo WebDriver
    driver = webdriver.Chrome()
    details = {}

    try:
        driver.get(url)
        
        # Chờ đợi phần tử <a> xuất hiện
        try:
            link = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'a.sc-d53bc9cf-1.dZxaOX'))
            )

            driver.execute_script("document.elementFromPoint(window.innerWidth/10, window.innerHeight/10).click();")
        except:
            pass

        try:    
            button = driver.find_element(By.XPATH, '//*[@id="detail-page"]/div[3]/div[2]/button/span')
            button.click()
        except:
            return None

        for _ in range(5):  # Lặp để cuộn nhiều lần
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                sleep(2)  # Đợi để trang tải xong

        soup = BeautifulSoup(driver.page_source)


        tmp = soup.find_all('li', class_ = 'sc-1f3f0799-1 sc-1f3f0799-3 eljeNi ePMbBG')


        for i in tmp:
            tmp_ = i.find_all('p')
            
            details[tmp_[0].text] = tmp_[1].text
            
    finally:
        driver.quit()
    
    return details
```

```python
data = []

# Đọc file và xử lý từng dòng
with open('DTGK_links.txt', 'r', encoding='utf-8') as file:
    for line in file:
        # Tách dòng thành sản phẩm và URL bằng dấu " : "
        parts = line.strip().split(' : ')
        if len(parts) == 2:  # Kiểm tra nếu dòng có đủ 2 phần
            product = parts[0]
            url = parts[1]
            data.append({"Product": product, "link": url})

# Tạo DataFrame từ danh sách dữ liệu
df_prod = pd.DataFrame(data)
df_prod.head()
```

```python
#Lấy giá
def get_price(url):
    driver = webdriver.Chrome()
    details = {}

    try:
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Tìm phần tử chứa giá
        price_element = soup.find('span', class_='sc-18fbacc2-2 eQBCtE')

        price_text = price_element.get_text(strip=True)

        # Làm sạch và chuyển đổi giá
        price_number = price_text.replace('đ', '').replace('.', '').strip()
        return price_number
    except:
        driver.quit()
        
get_price('https://dienthoaigiakho.vn/samsung-galaxy-a15-5g-8g256gb-chinh-hang-bhdt-1702886357')
```
