```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

from time import sleep

print('-- Finish importing package')
```

```python
def get_links_from_pages(start_page, end_page):
    driver = webdriver.Chrome()
    all_links = []

    try:
        for page_no in range(start_page, end_page + 1):
            driver.get(f'https://clickbuy.com.vn/dien-thoai?page={page_no}')
            sleep(2)  # Đợi trang tải xong
            
            try:
                WebDriverWait(driver, 11).until(
                    EC.presence_of_all_elements_located((By.CLASS_NAME, 'list-products__item'))
                )
            except:
                print(f"Warning: Page {page_no} may not have loaded all elements properly.")
            
            page_source = BeautifulSoup(driver.page_source, 'html.parser')
            samples = page_source.find_all('div', class_='list-products__item')
            
            for sample in samples:
                product = sample.find('a')
                link = product.get('href')
                title = product.get('title')
                if not link.startswith("https://clickbuy.com.vn/"):
                    link = "https://clickbuy.com.vn/" + link
                all_links.append((title, link))
                
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")
    
    finally:
        driver.quit()

    return list(set(all_links))


# Sử dụng hàm để lấy liên kết từ trang 1 đến trang 19
links = get_links_from_pages(1, 19)
print(links)
```

```python
def write_links_to_file(links, filename='ClickBuy_links.txt'):
    with open(filename, 'w', encoding='utf-8') as file:
        for title, link in links:
            title = title.strip()
            link = link.strip()
            file.write(f'{title} : {link}\n')
            
# Ghi các liên kết vào file
write_links_to_file(links)
print("Hoàn thành việc ghi các liên kết vào file.")
```

```python
def get_phone_details(url):
    # Khởi tạo WebDriver
    driver = webdriver.Chrome()
    details = {}

    try:
        driver.get(url)
        
        # Đợi cho bảng thông số kỹ thuật hoặc danh sách <li> xuất hiện
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.product-specification__content'))
        )
        # sleep(3)
        
        # click_ = driver.find_element(By.XPATH, '//*[@id="provinceList"]/li[1]')
        # click_.click()
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        try:
            click = driver.find_element(By.XPATH, '/html/body/main/div[2]/div/div/div[2]/div/div[4]/p[1]')
            click.click()
        except:
            pass
        
        # Lấy bảng thông số kỹ thuật nếu có
        try:
            table = driver.find_element(By.CSS_SELECTOR, 'div.product-specification__content')
            rows = table.find_elements(By.TAG_NAME, 'tr')
            
            for row in rows:
                key = row.find_element(By.TAG_NAME, 'th').text
                value = row.find_element(By.TAG_NAME, 'td').text
                
                details[key] = value
                    
        except:
            print("Không tìm thấy bảng thông số kỹ thuật.")
        
    finally:
        # Đóng trình điều khiển sau khi hoàn tất
        driver.quit()
    
    return details
```

```python
import pandas as pd

# Danh sách chứa thông tin chi tiết của từng sản phẩm
phone_details_list = []

# Giả sử `links` là danh sách các tuple (title, link) của sản phẩm
for title, link in links:
    print(f"Đang xử lý: {title}")
    details = get_phone_details(link)
    details['Product'] = title  # Thêm tên sản phẩm vào từ điển details
    phone_details_list.append(details)
    
# Tạo DataFrame từ danh sách thông tin chi tiết
df = pd.DataFrame(phone_details_list)
df.head()
```

```python
#thêm cột link vào data frame
tmp_links = []
for title, link in links:
    tmp_links.append(link)
    
df['link'] = tmp_links    
```

```python
df.to_csv('phone_details_.csv', index=False, encoding='utf-8')
```

```python
#Lấy giá
def get_price(url):
    driver = webdriver.Chrome()
    
    try:
        driver.get(url)
        
        page_source = BeautifulSoup(driver.page_source)
        
        price = page_source.find('div', class_ = 'product-price text-right')
        
        price_old = 0
        if price:
            price_old = price.find('p', class_ = 'price-old').text
        else:
            return None

    finally:
        driver.quit()
        
    if price_old != 0:
        # Loại bỏ chữ 'đ' ở cuối giá
        price_old = price_old.replace("đ", "").replace(",", "")
    
    return price_old
# Gọi hàm với URL cụ thể
# get_price('https://clickbuy.com.vn//tecno-pova-6')
```
